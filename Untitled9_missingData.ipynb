```python
import pandas as pd

df = pd.read_excel('/content/missingData.xlsx')

df
```

```python
df.info()
```

```python
df.head(100)
```

```python
#Removing rows with missing values
df.dropna()
```

```python
#Removing columns with missing values
df.dropna(axis=1)
```

```python
df.info()
```

```python
#Returns a dataFrame of the same shape with True for missing values
df.isnull()
```

```python
#Returns a dataFrame of the same shape with True for non-missing values
df.notnull()
```

```python
#Filling missing values with specified 'values'
df.fillna(1.07136)
```

```python
#Filling missing values with the previous non-missing values
df.fillna(method='ffill')
```

```python
#Replace missing values with a calcuated value, such as the mean or median.
df['close'].fillna(df['high'].mean(1.09149), inplace=True)
```
